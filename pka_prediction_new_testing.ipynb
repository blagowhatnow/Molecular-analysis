```python
import rdkit
from rdkit.Chem import PandasTools
```

```python
SDFFile1 = "mols.sdf"
BRDLigs1 = PandasTools.LoadSDF(SDFFile1)

SDFFile2 = "AvLiLuMoVe.sdf"
BRDLigs2 = PandasTools.LoadSDF(SDFFile2)

SDFFile3 = "combined_training_datasets_unique.sdf"
BRDLigs3 = PandasTools.LoadSDF(SDFFile3)

SDFFile4 = "novartis_cleaned_mono_unique_notraindata.sdf"
BRDLigs4 = PandasTools.LoadSDF(SDFFile4)

SDFFile5 = "chembl25.sdf"
BRDLigs5 = PandasTools.LoadSDF(SDFFile5)

SDFFile6 = "AvLiLuMoVe_cleaned_mono_unique_notraindata.sdf"
BRDLigs6 = PandasTools.LoadSDF(SDFFile6)

SDFFile7 = "datawarrior.sdf"
BRDLigs7 = PandasTools.LoadSDF(SDFFile5)
```

```python
pkas1=BRDLigs1['pka']
mols1=BRDLigs1['ROMol']

pkas2=BRDLigs2['pKa']
mols2=BRDLigs2['ROMol']

pkas3=BRDLigs3['pKa']
mols3=BRDLigs3['ROMol']

pkas4=BRDLigs4['pKa']
mols4=BRDLigs4['ROMol']

pkas5=BRDLigs5['pKa']
mols5=BRDLigs5['ROMol']

pkas6=BRDLigs6['pKa']
mols6=BRDLigs6['ROMol']

pkas7=BRDLigs6['pKa']
mols7=BRDLigs6['ROMol']
```

```python
import numpy as np 

pkas1=np.array([float(i) for i in pkas1])
pkas2=np.array([float(i) for i in pkas2])
pkas3=np.array([float(i) for i in pkas3])
pkas4=np.array([float(i) for i in pkas4])
pkas5=np.array([float(i) for i in pkas5])
pkas6=np.array([float(i) for i in pkas6])
pkas7=np.array([float(i) for i in pkas6])



pkas=np.concatenate([pkas1,pkas2,pkas3,pkas4,pkas5,pkas6,pkas7], axis=0)

mols=np.concatenate([mols1,mols2,mols3,mols4,mols5,mols6,mols7], axis=0)
```

```python
from rdkit import Chem
import numpy as np

smi=[]
p=[]

for i in np.arange(0,mols.shape[0]):
        try:
            sm = Chem.MolToSmiles(mols[i])
            smi.append(sm)
            p.append(pkas[i])
        except:
            pass
```

```python
#Remove duplicates

smi_unique = []
pka_unique = []

# Loop through elements of input lists
for a, b in zip(smi, p):
    # If it isn't a duplicate, append to the new lists
    if a not in smi_unique:
        smi_unique.append(a)
        pka_unique.append(b)
```

```python
smi_unique=np.array(smi_unique)
pka_unique=np.array(pka_unique)
```

```python
from rdkit import Chem
from rdkit.Chem import AllChem
from rdkit.Chem import DataStructs

m=[Chem.MolFromSmiles(i) for i in smi_unique] 


for mol in m:
    AllChem.Compute2DCoords(mol)
X = []
for mol in m:
    arr = np.zeros((0,))
    fp = AllChem.GetMorganFingerprintAsBitVect(mol, 2)
    DataStructs.ConvertToNumpyArray(fp, arr)
    X.append(arr)
```

```python
X=np.array(X)
y=np.array(pka_unique)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
from sklearn.ensemble import RandomForestRegressor

reg=RandomForestRegressor()
reg.fit(X_train,y_train)
pred_y=reg.predict(X_test)
```

```python
import joblib

filename = 'pka_model_new.sav'
joblib.dump(reg, filename)
```

```python
from sklearn.metrics import mean_squared_error
from math import sqrt

#calculate RMSE
sqrt(mean_squared_error(pred_y, y_test))
```
