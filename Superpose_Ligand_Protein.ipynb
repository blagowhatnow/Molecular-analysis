```python
from rdkit import Chem
from rdkit.Chem import AllChem

def split_sdf_file(sdf_path):
    """
    Split an SDF file into separate files for each molecule.
    Each file is named with consecutive numbers.

    Parameters
    ----------
    sdf_path: str or pathlib.Path
        Path to SDF file that should be split.
    """
    sdf_path = str(sdf_path)
    suppl = Chem.SDMolSupplier(sdf_path)

    for i, mol in enumerate(suppl, 1):
        if mol is not None:
            output_path = f"{sdf_path}_{i}.sdf"
            writer = Chem.SDWriter(output_path)
            writer.write(mol)
            writer.close()

# Example usage:
sdf_file_path = "selectedposeH.sdf"
split_sdf_file(sdf_file_path)
```
