```python
import rdkit
from rdkit.Chem import PandasTools
```

```python
SDFFile = "mols.sdf"
BRDLigs = PandasTools.LoadSDF(SDFFile)
```

```python
pkas=BRDLigs['pka']
mols=BRDLigs['ROMol']
```

```python
pkas=[float(i) for i in pkas]
```

```python
from rdkit import Chem
import numpy as np

smi=[]
p=[]

for i in np.arange(0,mols.shape[0]):
        try:
            sm = Chem.MolToSmiles(mols[i])
            smi.append(sm)
            p.append(pkas[i])
        except:
            pass
```

```python
smi=np.array(smi)
```

```python
p=np.array(p)
```

```python
smi.shape
```

```python
m=[Chem.MolFromSmiles(i) for i in smi] 

from rdkit import Chem
from rdkit.Chem import AllChem
from rdkit.Chem import Draw
from rdkit.Chem import DataStructs

for mol in m:
    AllChem.Compute2DCoords(mol)
X = []
for mol in m:
    arr = np.zeros((0,))
    fp = AllChem.GetMorganFingerprintAsBitVect(mol, 2)
    DataStructs.ConvertToNumpyArray(fp, arr)
    X.append(arr)
```

```python
np.array(m).shape
```

```python
X=np.array(X)
```

```python
X.shape
```

```python
from sklearn.ensemble import RandomForestRegressor

reg=RandomForestRegressor()
reg.fit(X,p)
pred_pkas=reg.predict(X)
```

```python
import joblib

filename = 'pka_model.sav'
joblib.dump(reg, filename)
```

```python
loaded_model = joblib.load('pka_model.sav')
```

```python
import seaborn as sns
from scipy import stats
def r2(x, y):
    return stats.pearsonr(x, y)[0] ** 2

r2(pred_pkas,p)
```
