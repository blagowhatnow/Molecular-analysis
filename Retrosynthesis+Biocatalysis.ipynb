```python
from transformers import pipeline
import numpy as np
from rdkit import Chem
from rdkit.Chem import AllChem
```

```python
num_beam=range(1,51)
```

```python
smiles='CC1(C2C1C(N(C2)C(=O)C(C(C)(C)C)NC(=O)C(F)(F)F)C(=O)NC(CC3CCNC3=O)C#N)C'
```

```python
preds=[]
count=0
for i in num_beam:
    pipe = pipeline("text2text-generation", model="sagawa/ReactionT5-retrosynthesis", num_beams=i, max_length=512)
    preds.append(pipe.predict(smiles))
    count+=1
    print("Done "+str(i)+ " out of "+str(num_beam[-1]+1))
```

```python
#Extract chemically valid reactions
valid_rxns=[]
for i in range(0,len(preds)):
  reactants=preds[i][0]['generated_text'].split('.')
  valid=[]
  for j in reactants:
      molecule=Chem.MolFromSmiles(j)
      if molecule is not None:
        valid.append(True)
      else:
        valid.append(False)
  if False not in valid:
    valid_rxns.append(preds[i][0]['generated_text'])
```

```python
#unique chemically valid reactions
valid_rxns=list(set(valid_rxns))
```

```python
#Total number of chemically valid reactions predicted by the model
len(valid_rxns)
```

```python
valid_rxns=[i.replace(' ','') for i in valid_rxns]  #Remove all unnecessary spaces
```

```python
#Visualize reactions
rxns=[]
for i in valid_rxns:
   rxn=AllChem.ReactionFromSmarts(smiles+">>"+i)
   rxns.append(rxn)
```

```python
rxns[0]
```

```python
#Reaction classification(See if a reaction predicted with high probability under any particular reaction class)
classes=[]
for i in valid_rxns:
    pipe = pipeline("text-classification", model="grabau/reaction-classification")
    res=pipe.predict(smiles+'>>'+i)
    classes.append(res)
```

```python
classes[0]
```

```python
#Generate an enzyme that could possibly catalyse the reaction. Generating just one for this case of a demo.
pipe = pipeline("text-generation", model="AI4PD/ZymCTRL")
res=pipe.predict(valid_rxns[0]+'>>'+smiles)
```

```python
predicted_enzyme=res[0]['generated_text'].replace(valid_rxns[0]+'>>'+smiles,"").replace(" ","")
```
